## 03_02 - Removendo Valores Repetidos

### Tipos de Imóveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados.head()
```

```python
tipo_de_imovel = dados['Tipo']  # Só os valores da variável 'Tipo'
```

```python
tipo_de_imovel
```

```python
type(tipo_de_imovel)  # Series (DataFrame é um conjunto de Series)
```

```python
tipo_de_imovel.drop_duplicates()  # mostra as variáveis sem repetição
```

```python
tipo_de_imovel  # não mudou nada
```

```python
# para mudar direto na variável "tipo_de_imovel" usamos o inplace=True (por padrão é False)
tipo_de_imovel.drop_duplicates(inplace=True)
```

```python
tipo_de_imovel
```

## 03_04 Redefinindo o Index

### Organizando a Visualização

```python
tipo_de_imovel = pd.DataFrame(tipo_de_imovel)  # transforamndo o Series em DataFrame
```

```python
tipo_de_imovel.index
```

```python
tipo_de_imovel.shape[0]  # tem 22 indexes 
```

```python
# Transoformando os indexes de 'tipo_de_imovel' em números em sequência da quantidade de valores no DataFrame 
tipo_de_imovel.index = range(tipo_de_imovel.shape[0])
```

```python
# Nomenado a coluna de index:
tipo_de_imovel.columns.name = 'Id'
```
