## 04_02 - Imóveis Residênciais

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Vendo quais tipos de aluguel existem
list(dados['Tipo'].drop_duplicates())
```

```python
# Variável com somente os Residenciais
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
residencial
```

```python
# Selecionando em True e False quais são os residenciais
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
# Criando outro DataFrame só com os imóveis Residenciais
alugueis_residenciais = dados[selecao]
alugueis_residenciais.head(10)
```

```python
# Conferindo se está somente os residenciais mesmo
list(alugueis_residenciais['Tipo'].drop_duplicates())
```

```python
# tamanho
alugueis_residenciais.shape[0]
```

```python
dados.shape[0]
```

```python
# Reconstruindo o index
alugueis_residenciais.index = range(alugueis_residenciais.shape[0])
```

```python
alugueis_residenciais
```

## 04_04 - Exportando Base de Dados

```python
# Exportando
alugueis_residenciais.to_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
# Importando o arquivo exportado
alugueis_residenciais_2 = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```
