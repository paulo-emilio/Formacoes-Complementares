# Anlisando as notas do Movielens

```python
notas = pd.read_csv('../dados/ratings.csv')
notas.head()
```

```python
nota_media_por_filme = notas.groupby('movieId').mean().rating
nota_media_por_filme.head()
```

```python
ax = sns.histplot(nota_media_por_filme.values, kde=True)
ax.grid()
ax.set(xlabel='Nota Média', 
       ylabel='Frequência', 
       title='Média de votos em filmes no Movielens 100k')

ax
```

```python
# contagem de votos por filme
qtd_votos_por_filme = notas.groupby('movieId').count()
# somente >= 10
notas_mais_de_10_votos = qtd_votos_por_filme.query('userId >= 10').index
notas_mais_de_10_votos.values
```

```python
# nota média por filmes que possuem 10 ou mais votos
nota_media_pelo_menos_10_votos = nota_media_por_filme.loc[notas_mais_de_10_votos.values]
nota_media_pelo_menos_10_votos.head()
```

```python
ax = sns.histplot(nota_media_pelo_menos_10_votos.values, kde=True)
ax.grid()
ax.set(xlabel='Nota Média', 
       ylabel='Frequência', 
       title='Média de votos em filmes no Movielens 100k (10 ou mais votos)')
ax
```

```python
ax = sns.boxplot(x=nota_media_pelo_menos_10_votos.values)
ax.set(xlabel='Nota média do filme', title='Média de votos em filmes no Movielens 100k (10 ou mais votos)')
ax
```

## Cumulative

```python
ax = sns.histplot(nota_media_pelo_menos_10_votos.values, 
                  kde=True, cumulative=True, stat="percent")
ax.grid()
ax.set(xlabel='Nota Média', 
       ylabel='% acumulada de filmes', 
       title='Média de votos em filmes no Movielens 100k (10 ou mais votos)')
ax
```

# ZTest para uma amostra

```python
filmes = pd.read_csv('../dados/movies.csv')
filmes.head()
```

```python
toystory = notas.query('movieId == 1')
toystory.head()
```

```python
ax = sns.histplot(toystory.rating, kde=True)
ax.grid()
ax.set(title='Distribuição de notas de Toy Story', xlabel='Notas', ylabel='Quantidade')
ax
```

```python
ax = sns.boxplot(x=toystory.rating)
ax.grid(axis='x')
ax.set(title='Box-plot de notas de Toy Story', xlabel='Notas')
ax
```

```python
toystory.rating.mean().round(2)
```

```python
# zconfint

testez_toy = zconfint(toystory.rating)
print(f'No mundo real, média entre: {testez_toy[0].round(2)} e {testez_toy[1].round(2)}')
```

```python
# comparando com a média geral 
# no mundo real, a média de toystory é igual a médias dos filmes em geral?

pvalue = ztest(toystory.rating, value=nota_media_pelo_menos_10_votos.mean())[1]
if pvalue < 0.05:
    print('Rejeitamos a hipótese nula, a média de Toy Story não é igual à dos filmes em geral.')
else:
    print('Não podemos rejeitar a hipótese nula, a média de Toy Story é igual à dos filmes em geral.')
```

```python
# Média de Toy Story a medida que aumentam as notas

# criando ordem aleatória
np.random.seed(75241)
temp = toystory.sample(frac=1).rating

# lista
medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```

```python
# Vendo o zteste alterando 
# Média de Toy Story a medida que aumentam as notas

# criando ordem aleatória
np.random.seed(75241)
temp = toystory.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value=nota_media_pelo_menos_10_votos.mean())
    return (i, media, p)

# lista
valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.plot(valores[:, 0], valores[:, 1])
plt.plot(valores[:, 0], valores[:, 2])

# linha no 0.05 para ver se passou no teste ou nao
plt.hlines(y=0.05, xmin=0, xmax=len(temp), colors='k')
```

### Perebemos que em alguns pontos as médias podem ser iguais (p value > 0.05), mas nesse caso teremos que concluir que não há dados suficientes para tirar essa conclusão

* que não há evidências para descartar a hipótese nula
* que não há dados suficientes para tirar essa conclusão

Toy Story tem notas 0.28 à 0.56 mais altas que os filmes em geral.

```python
# T-test
from scipy.stats import ttest_ind

ttest_ind(toystory.rating, notas.rating)
```

# Fizemos todos esses testes mas eles devem ser feitos se a distruibuição dos dados for normal, e não verificamos isso, então vamos verificar

```python
from scipy.stats import normaltest

_, p = normaltest(toystory.rating)
p
```

p < 0.05, logo, descartamos a hipótese nula. 

Os dados não possuem uma destribuição normal. 

# Não são normais, então devemos usar testes "Não paramétricos"

## ranksums

```python
from scipy.stats import ranksums

# Toy Story vs The Silence of the Lambs
_, p = ranksums(toystory.rating, tsotl.rating)
p

# hipótese nula => as duas amostras vieram da mesma distribuição
# hipótese alternativa => valores de uma amostra tem a tendência de serem maiores que da outra
```

### p menor 0.05 --> Descartamos a hipótese nula. 
#### Valores de uma amostra tem a tendência de serem maiores que da outra.
#### Pegando uma nota aleatória de Toy Story, e um de The Silence of the Lambs, provavelmente não teremos notas iguais.
