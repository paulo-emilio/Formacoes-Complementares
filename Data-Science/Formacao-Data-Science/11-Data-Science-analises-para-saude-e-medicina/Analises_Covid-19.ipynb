# Estudo de casos, mortes e recuperados por Covid-19 '8/4/21'

```python
# Data
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

# fatiamento do dataframe até '8/4/21' (último dia que foi atualizado)
confirmed = confirmed.loc[:, :'8/4/21']
deaths = deaths.loc[:, :'8/4/21']
recovered = recovered.loc[:, :'8/4/21']

display(recovered.head(2))
display(deaths.head(2))
display(confirmed.head(2))
```

```python
def latest_by_country_at(data, date):
    return data.groupby('Country/Region').sum(numeric_only=True)[date]
```

```python
latest_by_country_at(deaths, '8/4/21').head()  # teste
```

```python
# Juntando dados
all_datas = [latest_by_country_at(confirmed, '8/4/21'), 
             latest_by_country_at(deaths, '8/4/21'), 
             latest_by_country_at(recovered, '8/4/21')]

combined = pd.concat(all_datas, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']

combined.head()
```

```python
# Taxa de letalidade global em 8/4/21

sum_up = combined.sum()

global_letality_1 = sum_up.deaths / sum_up.confirmed
global_letality_2 = sum_up.deaths / (sum_up.deaths + sum_up.recovered)

print(f'Letalidade (mortes / confirmados): {(global_letality_1 * 100).round(2)} %\n')
print(f'Letalidade (mortes / mortes+recuperados): {(global_letality_2 * 100).round(2)} %')
```

```python
# Taxa de letalidade por país em 8/4/21

# letalidade 1 (deaths / confirmed)
letality_rate_1 = combined.deaths / combined.confirmed

# letalidade 2 (deaths / (deaths + recovered))
letality_rate_2 = combined.deaths / (combined.deaths + combined.recovered)

display(letality_rate_1.head())
display(letality_rate_2.head())
```

```python
# adicionando ao DataFrame
combined['letality_rate_1'] = letality_rate_1 * 100
combined['letality_rate_2'] = letality_rate_2 * 100

combined.head()
```

```python
# Ordenando por letalidade
combined.sort_values('letality_rate_1', ascending=False).head(10)
```

### Plotando dados

```python
# Horrível
combined[['confirmed', 'letality_rate_1']].plot()
```

```python
import seaborn as sns
combined['confirmed_in_millions'] = combined['confirmed'] / 1e6

sns.scatterplot(data=combined, x='confirmed_in_millions', y='letality_rate_1')
```

```python
# somente mais de 40 casos
sns.scatterplot(data = combined.query('confirmed_in_millions > 40/1e6 and confirmed_in_millions < 7'), 
                x = 'confirmed_in_millions', y = 'letality_rate_1')
```

```python
# letality rate 2
sns.scatterplot(data = combined.query('recovered > 40 and confirmed_in_millions < 7'), 
                x = 'confirmed_in_millions', y = 'letality_rate_2')
```

```python
# Histograma
sns.histplot(combined.query('confirmed > 40').letality_rate_1)
```

# Comparações e Análises

### Comparando Relatórios

```python
# Dados dia 2/20/20

date = '2/20/20'
country = 'China'

all_datas = [latest_by_country_at(confirmed, date), 
             latest_by_country_at(deaths, date), 
             latest_by_country_at(recovered, date)]

combined = pd.concat(all_datas, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']

country_on_date = combined.loc[country]
letality_rate_1 = country_on_date.deaths / country_on_date.confirmed
letality_rate_2 = country_on_date.deaths / (country_on_date.deaths + country_on_date.recovered)

display(letality_rate_1 * 100)
display(letality_rate_2 * 100)
```
