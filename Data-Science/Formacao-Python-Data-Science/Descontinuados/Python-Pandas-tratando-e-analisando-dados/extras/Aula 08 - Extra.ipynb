## Cirando Faixas de Valor

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head()
```

```python
# Contar imóveis com:
# 1 e 2
# 2 e 4
# 5 e 6
# 7 +
classes = [0, 2, 4, 6, 100] # primeiro limite 2, proximo limite 4, proximo 6, proximo 100
```

```python
quartos = pd.cut(dados.Quartos, classes)
quartos
```

```python
# Contagem 
pd.value_counts(quartos)
# o '(' significa que o número que está depois não entra na conta, e o ']' significa que entra
# ou seja (0, 2] -> o '0' não entra, o '2' entra
```

```python
labels = ['1 e 2 quartos', '3 e 4 quartos', '5 e 6 quartos', '7 ou mais quartos']
```

```python
# colocando labels
quartos = pd.cut(dados['Quartos'], classes, labels=labels)
pd.value_counts(quartos)
```

```python
# incluindo o 0 no '1 a 2'
labels = ['0 à 2 quartos', '3 e 4 quartos', '5 e 6 quartos', '7 ou mais quartos']
quartos = pd.cut(dados['Quartos'], classes, labels=labels, include_lowest=True)
pd.value_counts(quartos)
```
